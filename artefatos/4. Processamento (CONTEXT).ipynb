```
import pyspark.sql.functions as fn

from pyspark.sql import SQLContext
sqlContext = SQLContext(sc)
```

##1. Prepara camada context

```
df_marca = sqlContext.sql("""
SELECT marca, count(marca) as total, ((count(marca)/(SELECT count(marca) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
and tipo_veiculo = 'AutomÃ³vel'
GROUP BY marca
ORDER BY 2 desc
"""
)

df_marca.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.marca', path='/FileStore/context/marca')
```

```
df_uf = sqlContext.sql("""
SELECT uf, count(uf) as total, ((count(uf)/(SELECT count(uf) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
GROUP BY uf
ORDER BY count(uf) DESC
"""
)

df_uf.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.uf', path='/FileStore/context/uf')
```

```
df_sexo = sqlContext.sql("""
SELECT sexo, count(sexo) as total, ((count(sexo)/(SELECT count(sexo) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
GROUP BY sexo
ORDER BY count(sexo) DESC
"""
)

df_sexo.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.sexo', path='/FileStore/context/sexo')
```

```
df_idade = sqlContext.sql("""
SELECT idade, count(idade) as total, ((count(idade)/(SELECT count(idade) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
and idade >= 18 and idade < 100
GROUP BY idade
ORDER BY 1
"""
)

df_idade.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.idade', path='/FileStore/context/idade')
```

```
df_ano_fabricacao = sqlContext.sql("""
SELECT ano_fabricacao_veiculo, count(ano_fabricacao_veiculo) as total, ((count(ano_fabricacao_veiculo)/(SELECT count(ano_fabricacao_veiculo) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
GROUP BY ano_fabricacao_veiculo
ORDER BY 2 desc
"""
)

df_ano_fabricacao.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.ano_fabricacao', path='/FileStore/context/ano_fabricacao')
```

```
df_tipo_veiculo = sqlContext.sql("""
SELECT tipo_veiculo, count(tipo_veiculo) as total, ((count(tipo_veiculo)/(SELECT count(tipo_veiculo) FROM raw.acidentes)) * 100) percentual
FROM raw.acidentes
where tipo_envolvido = 'Condutor'
GROUP BY tipo_veiculo
ORDER BY 2 desc
"""
)

df_tipo_veiculo.write.format('delta').option("encoding", "ISO-8859-1").mode('overwrite').saveAsTable('context.tipo_veiculo', path='/FileStore/context/tipo_veiculo')
```
